```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
salario = np.array([100,200,300,500,400,150])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c='green')
```

```python
plt.plot(salario, c='green', ls='--')
```

```python
plt.plot(salario, ls='--', marker='s')
```

```python
plt.plot(salario,c='green',ls='--',marker='s',ms=19)
plt.show()
```

```python
salarios_marcos = np.array([100,200,400,300])
salarios_gileno = np.array([50,300,500,450])
```

```python
plt.plot(salarios_marcos,c='blue',ls='--',marker='s',ms=8,label='Salrios do Marcos')
plt.plot(salarios_gileno,c='red',ls='--',marker='s',ms=8,label='Salrios do Gileno')
plt.show()
```

```python
plt.plot(salarios_marcos,c='blue',ls='--',marker='s',ms=8,label='Salrios do Marcos')
plt.plot(salarios_gileno,c='red',ls='--',marker='s',ms=8,label='Salrios do Gileno')
plt.legend()
plt.show()
```

```python
plt.plot(salarios_marcos,c='blue',ls='--',marker='s',ms=8,label='Salrios do Marcos')
plt.plot(salarios_gileno,c='red',ls='--',marker='^',ms=8,label='Salrios do Gileno')
plt.legend()
plt.show()
```

```python
plt.plot(salarios_marcos,c='blue',ls='--',marker='s',ms=8,label='Salrios do Marcos')
plt.plot(salarios_gileno,c='red',ls='--',marker='^',ms=8,label='Salrios do Gileno')
plt.legend(loc='lower right')
plt.show()
```

```python
salarios_pedro = np.array([200,150,300,700])
plt.plot(salarios_marcos,c='blue',ls='--',marker='s',ms=8,label='Salrios do Marcos')
plt.plot(salarios_gileno,c='red',ls='--',marker='^',ms=8,label='Salrios do Gileno')
plt.plot(salarios_pedro,c='green',ls='--',marker='o',ms=10,label='Salrios do Pedro')
plt.legend(loc='lower right')
plt.show()
```
